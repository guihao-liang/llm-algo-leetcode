# Transforms

## cat vs stack

- cat on existing dim
  - shape same except the cat dim
- stack on new dim
  - shape must be same

```python
# b, nh, s, hd
xk1 = torch.randn(1, 2, 4, 10)
xk2 = torch.randn(1, 2, 3, 10)

assert torch.cat([xk1, xk2], dim = 2).shape[2] == 7
```

```python
# list of tensors -> dense tensors repr
xk1 = torch.randn(1, 3, 4, 10)
xk2 = torch.randn(1, 3, 4, 10)

print(torch.stack((xk1, xk2), dim=-1).shape)
torch.stack((xk1, xk2), dim=0).shape
```

## expand

- only on dim 1
- map the new dim system using the old one

```python
# 1. 准备原始张量 (Shape: [3, 1, 4, 1])
x = torch.randn(3, 1, 4, 1)

# 目标维度参数
dim0, dim1_target, dim2, dim3_target = 3, 5, 4, 7

# 2. 显式开辟全新的目标显存空间 (Shape: [3, 5, 4, 7])
out = torch.zeros((dim0, dim1_target, dim2, dim3_target), dtype=x.dtype)

# 3. 嵌套 for 循环进行物理赋值
for d0 in range(dim0):                  # 遍历第 0 维 (大小 3)
    for d1 in range(dim1_target):       # 遍历第 1 维 (大小 5)
        for d2 in range(dim2):          # 遍历第 2 维 (大小 4)
            for d3 in range(dim3_target): # 遍历第 3 维 (大小 7)
                
                # 核心魔法在这里：
                # d1 和 d3 在源张量中对应的维度大小只有 1，所以读取时索引固定为 0
                out[d0, d1, d2, d3] = x[d0, 0, d2, 0]

# 验证结果是否与原生 expand 一致
x_expand = x.expand(3, 5, 4, 7)
print("结果是否完全一致:", torch.allclose(out, x_expand)) # 输出: True
```
